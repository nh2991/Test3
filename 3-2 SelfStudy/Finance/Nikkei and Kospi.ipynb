```python
Nikkei225_diff = []

for i in np.arange(1,len(Nikkei225.index)-1):
    Nikkei225_diff.append((Nikkei225['Adj Close'][i] - Nikkei225['Adj Close'][i-1]) / Nikkei225['Adj Close'][i])
```

```python
index_diff = Nikkei225.index[1:457]
```

```python
Nikkei225_diff = pd.DataFrame(Nikkei225_diff, index=index_diff)
```

```python
plt.plot(Nikkei225_diff)
```

```python
year_diff = index_diff[[0, 50, 100, 150, 200, 250, 300, 350, 400, 450]]
```

```python
plt.figure(figsize=(16,12))
plt.plot(Nikkei225_diff)
plt.xticks(year_diff)
plt.show()
```

```python
Kospi_diff = []

for i in np.arange(1,len(Kospi.index)-1):
    Kospi_diff.append((Kospi['Adj Close'][i] - Kospi['Adj Close'][i-1]) / Kospi['Adj Close'][i])
```

```python
index_diff_Kospi = Kospi.index[1:313]
```

```python
Kospi_diff = pd.DataFrame(Kospi_diff, index=index_diff_Kospi)
Kospi_diff
```

```python
plt.figure(figsize=(16,12))
plt.plot(Nikkei225_diff, color='blue', alpha=0.5, label='Nikkei')
plt.plot(Kospi_diff, color='red', alpha=0.5, label='Kospi')
plt.legend(fontsize=12)
plt.xticks(year_diff)
plt.title("Percentage Change of Nikkei and Kospi, 85-22 and 97-22", fontsize=15)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Perchantage Change", fontsize=12)
plt.show()
```

```python
Nikkei225_logdiff = []

for i in np.arange(1,len(Nikkei225.index)-1):
    a = np.log(Nikkei225['Adj Close'][i])
    b = np.log(Nikkei225['Adj Close'][i-1])
    Nikkei225_logdiff.append(a-b)
```

```python
Nikkei225_logdiff
```

```python
Nikkei225_logdiff = pd.DataFrame(Nikkei225_logdiff, index=index_diff)
```

```python
Nikkei225_logdiff
```

```python
Kospi_logdiff = []

for i in np.arange(1,len(Kospi.index)-1):
    a = np.log(Kospi['Adj Close'][i])
    b = np.log(Kospi['Adj Close'][i-1])
    Kospi_logdiff.append(a-b)
```

```python
Kospi_logdiff = pd.DataFrame(Kospi_logdiff, index=index_diff_Kospi)
```

```python
Kospi_logdiff
```

```python
fig = plt.figure(figsize=(16,12))
plt.plot(Nikkei225_logdiff, color='blue', alpha=0.5, label='Nikkei 225')
plt.plot(Kospi_logdiff, color='red', alpha=0.5, label='Kospi')
plt.legend(loc='upper right', fontsize=15, ncol=2)
plt.xticks(year_diff)

plt.title("Log-Scale Percentage Change of Nikkei 225 and Kospi, 85-22 and 97-22", fontsize=20)
plt.xlabel("Year", fontsize=15)
plt.ylabel("Log-Scale Perchantage Change", fontsize=15)

plt.axvline(x='1986-01-01', color='black', alpha=0.1, linestyle='-', linewidth=10)
plt.annotate('Plaza Agreement, \n85/09/22', xy=(0,0.2), xytext=(0,0.18), fontsize=12, color='black')
plt.axvline(x='1993-03-01', color='black', alpha=0.1, linestyle='-', linewidth=120)
plt.annotate('Bubble Burst, \n90-95$^*$', xy=(75,-0.25), xytext=(75,-0.235), fontsize=12, color='black')
plt.axvline(x='1999-08-01', color='black', alpha=0.1, linestyle='-', linewidth=80)
plt.annotate('Asian Financial Crisis, \n97-01$^*$', xy=(150,-0.3), xytext=(160,-0.3), fontsize=12, color='black')
plt.axvline(x='2009-08-01', color='black', alpha=0.1, linestyle='-', linewidth=60)
plt.annotate('Sub-Prime Mortgage Crisis, \n07-09$^*$', xy=(300, -0.25), xytext=(290, -0.25), fontsize=12, color='black')
plt.axvline(x='2021-08-01', color='black', alpha=0.1, linestyle='-', linewidth=60)
plt.annotate('COVID-19, \n19-$^*$', xy=(400,0.15), xytext=(420,0.175), fontsize=12, color='black')

plt.show()
```
