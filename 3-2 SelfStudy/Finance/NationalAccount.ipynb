```python
fig4 = plt.figure(figsize=(16,12))
plt.plot(NaAccount[['소비자물가지수 변동률']], label='Consumer Price Index Rate', color='r')
plt.plot(NaAccount[['생산자물가지수 변동률']], label='Producer Price Index Rate', color='orange')
plt.plot(NaAccount[['1인당 국내총생산 변동률']], label='GDP per capita Rate (Nominal, Won)', color='y')
plt.plot(NaAccount[['1인당 국민총소득 변동률']], label='GNI per capita Rate (Nominal, Won)', color='green')
plt.plot(NaAccount[['국내총생산 변동률']], label='GDP Rate (Nominal, Won)', color='blue')
plt.plot(NaAccount[['수입물가지수 변동률']], label='Import Price Index Rate', color='navy')
plt.plot(NaAccount[['수출물가지수 변동률']], label='Export Price Index Rate', color='purple')
plt.xticks(rotation=90, fontsize=10)
plt.legend()
plt.show()
```

```python
fig5 = plt.figure(figsize=(16,12))
plt.plot(NaAccount[['생산자물가지수 변동률']], label='Producer Price Index Rate', color='green')
plt.plot(NaAccount[['수입물가지수 변동률']], label='Import Price Index Rate', color='blue', alpha=0.666)
plt.plot(NaAccount[['수출물가지수 변동률']], label='Export Price Index Rate', color='red', alpha=0.666)
plt.xticks(rotation=90, fontsize=12)
plt.legend()
plt.title('Various Price Index of Republic of Korea, 1971-2021', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('So-called Inflation', fontsize=12)
plt.axvline(x=9, color='red', linestyle='-', linewidth=20, alpha=0.1)
plt.axvline(x=3, color='red', linestyle='-', linewidth=20, alpha=0.1)
plt.axvline(x=27, color='blue', linestyle='-', linewidth=40, alpha=0.1)
plt.axvline(x=37, color='green', linestyle='-', linewidth=20, alpha=0.1)
plt.axvline(x=52, color='grey', linestyle='-', linewidth=100, alpha=0.1)
plt.annotate('1st Oil Shock, 73-74', xy=(3,1.44), xytext=(-1,1.44), fontsize=12, color='black')
plt.annotate('2nd Oil Shock, 79-81', xy=(7,1.6), xytext=(5, 1.6), fontsize=12, color='black')
plt.annotate('IMF, 97-01', xy=(27, 1.35), xytext=(25, 1.33), fontsize=12, color='black')
plt.annotate('Sub-prime Mortgage, 07-08', xy=(37, 1.35), xytext=(32, 1.38), fontsize=12, color='black')
plt.annotate('COVID-19, 20-', xy=(50, 1.2), xytext=(46, 1.2), fontsize=12, color='black')
plt.show()
```
