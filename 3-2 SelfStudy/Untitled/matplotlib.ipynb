```python
label = ['Thur', 'Fri', 'Sat', 'Sun']
index = np.arange(len(label))
index
```

```python
plt.figure(figsize=(5,3))
plt.bar(index, tips_sum_by_day)
plt.title('Sum of Tips by Day', fontsize=12)
plt.xlabel('Day', fontsize=10)
plt.ylabel('Sum of Tips', fontsize=10)
plt.xticks(index, label, fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar(index, tips_sum_by_day, color='red', alpha=0.5)
plt.title('Sum of Tips by Day', fontsize=12)
plt.xlabel('Day', fontsize=10)
plt.ylabel('Sum of Tips', fontsize=10)
plt.xticks(index, label, fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar(label, tips_sum_by_day, width=0.3)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.barh(index, tips_sum_by_day)
plt.title('Sum of Tips by Day', fontsize=12)
plt.ylabel('Day', fontsize=10)
plt.xlabel('Sum of Days', fontsize=10)
plt.yticks(index, label, fontsize=10)
plt.show()
```

```python
tips_sum_by_day_male = tips[tips['sex'] == 'Male'].groupby('day').tip.sum()
tips_sum_by_day_female = tips[tips['sex'] == 'Female'].groupby('day').tip.sum()
print('--Male--')
print(tips_sum_by_day_male)
print(' ')
print('--Female--')
print(tips_sum_by_day_female)
```

```python
label = ['Thur', 'Fri', 'Sat', 'Sun']
N = len(tips['day'].unique())
index = np.arange(N)
index
```

```python
plt.figure(figsize=(5,3))
p1 = plt.bar(index, tips_sum_by_day_male, color='b', alpha=0.5)
p2 = plt.bar(index, tips_sum_by_day_female, color='r', alpha=0.5,
            bottom = tips_sum_by_day_male)
plt.title('Stacked Bar Chart of Sum of Tips by Day & Sex', fontsize=12)
plt.ylabel('Sum of Tips', fontsize=10)
plt.xlabel('Day', fontsize=10)
plt.xticks(index, label, fontsize=10)
plt.legend((p1, p2), ('Male', 'Female'), fontsize=10)
plt.show()
```

 (Contour Plot) 3   .

```python
def my_func(x,y):
    z = np.sqrt(2* x**2 + 6* y**2)
    return np.sin(z) / z
```

```python
points = np.linspace(-8, 8, 50)
points
```

```python
XX, YY = np.meshgrid(points, points)
```

```python
XX
```

```python
YY
```

```python
ZZ = my_func(XX, YY)
```

```python
plt.contour(XX, YY, ZZ)
plt.show()
```

```python
plt.contourf(XX, YY, ZZ, alpha=0.5)
plt.show()
```
